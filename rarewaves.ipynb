```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./rarewaves_data.csv")
df
```

```python
df['Review Language'].value_counts()
```

```python
df_intermediatary = df[df['Review Language'] == 'en']
```

```python
df_intermediatary['Review Language'].value_counts()
```

```python
df_intermediatary
```

```python
df_intermediatary.describe()
```

```python
df_intermediatary = df_intermediatary.drop(['Review Id', 'Review Created (UTC)', 'Review Consumer User Id', 'Review Username', 'Source Of Review', 'Company Response', 'Review Language', 'Webshop Name'], axis=1)
```

```python
df_intermediatary
```

```python
df_intermediatary['good_review'] = df_intermediatary['Review Stars'] > 3
```

```python
df_intermediatary
```

```python
df_good_reviews = df_intermediatary[df_intermediatary['good_review'] == True]
```

```python
df_good_reviews
```

```python
df_good_reviews['Review Content']
```

```python
good_reviews_review_content = df_good_reviews['Review Content']
```

```python
good_reviews_review_content_lowercase = good_reviews_review_content.str.lower()
```

```python
good_reviews_review_content_lowercase
```

```python
good_reviews_review_content_lowercase_no_punctuations = good_reviews_review_content_lowercase.str.replace('[^\w\s]','')
```

```python
good_reviews_review_content_lowercase_no_punctuations
```

```python
good_reviews_review_content_lowercase_no_punctuations_list = good_reviews_review_content_lowercase_no_punctuations.tolist()
```

```python
good_reviews_review_content_lowercase_no_punctuations_list
```

```python
good_reviews_review_content_lowercase_no_punctuations_space_split_list = []

for element in good_reviews_review_content_lowercase_no_punctuations_list:
    good_reviews_review_content_lowercase_no_punctuations_space_split_list.append(element.split())
```

```python
good_reviews_review_content_lowercase_no_punctuations_space_split_list
```

```python
good_reviews_review_content_lowercase_no_punctuations_space_split_list_min_four_len = []

for element in good_reviews_review_content_lowercase_no_punctuations_space_split_list:
    good_reviews_review_content_lowercase_no_punctuations_space_split_list_min_four_len.append([word for word in element if len(word) > 3])
```

```python
good_reviews_review_content_lowercase_no_punctuations_space_split_list_min_four_len
```

```python
good_reviews_review_content_lowercase_no_punctuations_space_split_list_min_four_len_flat = [element for sublist in good_reviews_review_content_lowercase_no_punctuations_space_split_list_min_four_len for element in sublist]
```

```python
good_reviews_review_content_lowercase_no_punctuations_space_split_list_min_four_len_flat
```

```python
for element in good_reviews_review_content_lowercase_no_punctuations_space_split_list_min_four_len_flat:
    print(element)
```

```python
df = pd.DataFrame(good_reviews_review_content_lowercase_no_punctuations_space_split_list_min_four_len_flat)
df.to_clipboard(index=False,header=False)
```

```python
good_reviews_review_content_lowercase_no_punctuations_space_split_list_min_four_len_series = pd.Series(good_reviews_review_content_lowercase_no_punctuations_space_split_list_min_four_len_flat)
```

```python
good_reviews_review_content_lowercase_no_punctuations_space_split_list_min_four_len_series
```

```python
good_reviews_review_content_lowercase_no_punctuations_space_split_list_min_four_len_series_value_counts = good_reviews_review_content_lowercase_no_punctuations_space_split_list_min_four_len_series.value_counts()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

#convert list to string and generate
unique_string=(" ").join(good_reviews_review_content_lowercase_no_punctuations_space_split_list_min_four_len_flat)
wordcloud = WordCloud(width = 1000, height = 500).generate(unique_string)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig("word_cloud_gen_1"+".png", bbox_inches='tight')
plt.show()
plt.close()
```
